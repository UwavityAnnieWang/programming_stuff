```python
import matplotlib.pyplot as plt  # import pyplot submodule from matplotlib
import numpy as np               # also import numpy
import pandas as pd
```

```python
df = pd.read_csv('~/Downloads/heart.csv')
print(df.columns)
df.head()
```

```python
results = np.array(df).tolist()
ages = list(set([thing[0] for thing in results]))
print(ages)
count = np.array([[age, [thing[0] for thing in results].count(age)] for age in ages], dtype = object)
dictionary_ages = dict(count)
#print(dictionary_ages)


```

```python
thing = [[x[4] for x in results if x[0]==bruh ]for bruh in ages]
print(thing)
```

```python
average = [[(sum(thing[bruh]))/dictionary_ages[ages[bruh]]]for bruh in range(len(thing))]
print(average)

#sum(x[12] for x in total_debt_less
    
```

```python
plt.figure(figsize=(20, 10))
plt.plot(ages, average)
plt.xlabel("Age")
plt.ylabel("Average Cholesterol")
plt.title("Average Cholesterol Levels in Relation to Age")
```

```python
heartRate = [[x[7] for x in results if x[0]==bruh ]for bruh in ages]
heartRateAverage = [[(sum(heartRate[bruh]))/dictionary_ages[ages[bruh]]]for bruh in range(len(thing))]
plt.figure(figsize=(20, 10))
plt.plot(ages, heartRateAverage)
plt.xlabel("Age")
plt.ylabel("Average Max Heart Rate")
plt.title("Average Max Heart Rate in Relation to Age")
```

```python
restingBP = [[x[3] for x in results if x[0]==bruh ]for bruh in ages]
restingBPAverage = [[(sum(restingBP[bruh]))/dictionary_ages[ages[bruh]]]for bruh in range(len(thing))]
plt.figure(figsize=(20, 10))
plt.plot(ages, restingBPAverage)
plt.xlabel("Age")
plt.ylabel("Average Resting BP")
plt.title("Average Resting BP in Relation to Age ")

```
